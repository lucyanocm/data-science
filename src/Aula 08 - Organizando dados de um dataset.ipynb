```python
def tratarNome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\3 \1"', linha)
```

```python
# imprimir lista de passageiros com nome 'tratado'

for passageiro in linhasTitanic[1:]:
    print(tratarNome(passageiro))
```

```python
# Tratar todas as linhas excluíndo a linha 0 que é a descrição das colunas.

for dado in linhasTitanic[1:]:
    dado_trato = tratarNome(dado)
    dado_como_lista = dado_trato.split(',')
    print(dado_como_lista)
```

```python
dados_tratados = []

for dado in linhasTitanic[1:]:
    dado_trato = tratarNome(dado)
    dado_como_lista = dado_trato.split(',')
    dados_tratados.append(dado_como_lista)
```
